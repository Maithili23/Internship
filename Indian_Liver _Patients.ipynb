```python
from sklearn.model_selection import train_test_split as tts #cross_validation
from sklearn.linear_model import LogisticRegression as logreg
```

```python
#Test the model
# 1. Predict the output foe x_test data. Use predict function, pass x_test as arg
# 2. Check the accuracy by giving x_test and y_test data

# 1. predict o/p
y_pred=lr.predict(x_test)
y_pred
```

```python
# 2. checking accuracy
# Using score(x_test, y_test)
accuracy=lr.score(x_test,y_test)
print('Logistic Regression Accuracy: ',accuracy)
```

## KNN model

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
y=label.fit_transform(y)
y
```

```python
#Split the universal dataset
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.8,random_state=99)
```

```python
from sklearn.neighbors import KNeighborsClassifier as knn
knnmodel=knn(n_neighbors=95) 
```

```python
#Different k values, different accuracies
k_range=range(1,96)
k_score=[]
k_best =1
acc_best=0

for k in k_range:
    knnmodel=knn(n_neighbors=k)
    knnmodel.fit(x_train,y_train)
    k_acc=knnmodel.score(x_test,y_test)
    k_score.append(k_acc)
    
    if(acc_best<k_acc):
        acc_best= k_acc
        k_best  = k
```

```python
k_best
```

```python
k_acc
```

```python
plt.figure(figsize=(10,7))
plt.plot(k_range,k_score)
plt.xlabel("K values")
plt.ylabel("Accuracy")
plt.title("K valuse vs Accuracy")
plt.plot()
```

```python
#train x_train,y_train
knnmodel.fit(x_train,y_train)
```

```python
#test x_test, y_test
accuracy=knnmodel.score(x_test, y_test)
pred=knnmodel.predict(x_test)
```
